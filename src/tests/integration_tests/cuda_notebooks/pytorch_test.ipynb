# PyTorch CUDA smoke test

```python
import torch

print('PyTorch version:', torch.__version__)
print('CUDA available:', torch.cuda.is_available())
if not torch.cuda.is_available():
    raise RuntimeError('CUDA is not available inside the container')

device = torch.device('cuda')
x = torch.tensor([1.0, 2.0, 3.0], device=device)
y = torch.tensor([4.0, 5.0, 6.0], device=device)
z = x + y
assert torch.allclose(z, torch.tensor([5.0, 7.0, 9.0], device=device))
print('Vector add ok:', z)

a = torch.randn((2, 3), device=device)
b = torch.randn((3, 4), device=device)
c = a @ b
assert c.shape == (2, 4)
print('Matmul ok, result shape:', c.shape)
```

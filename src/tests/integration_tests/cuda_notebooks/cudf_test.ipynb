# cuDF smoke test

```python
import cudf
import numpy as np

# Basic groupby aggregation
df = cudf.DataFrame({'id': [1, 1, 2], 'value': [10, 20, 30]})
result = df.groupby('id').value.sum()
assert result.to_pandas().tolist() == [30, 30], result
print('Groupby sum ok:', result.to_pandas())

# Merge correctness
left = cudf.DataFrame({'k': [1, 2], 'v_left': ['a', 'b']})
right = cudf.DataFrame({'k': [2, 3], 'v_right': ['x', 'y']})
merged = left.merge(right, on='k', how='left')
assert merged.loc[merged.k == 2, 'v_right'].iloc[0] == 'x'
print('Merge ok:\n', merged)

print('cuDF version:', cudf.__version__)
```
